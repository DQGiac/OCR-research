```python
from pdf2image import convert_from_path
from PIL import Image
pages = 0
boxes = 6
box_regions = [[60, 200, 350, 240], [430, 240, 780, 280], [45, 280, 990, 320], [110, 320, 750, 480], [50, 660, 990, 800]]

def convert_pdf_to_images(pdf_path):
    pdf = convert_from_path(pdf_path, fmt='png')

    for i in range(len(pdf)):
        global pages
        pages += 1
        # pdf[i].save(f"og{i}.png")
        # page = Image.open("og0.png")
        page = pdf[i]
        for j in range(boxes):
            page_box = page.crop(box_regions[j])
            page_box.save(f"./pages/page_{i}_box_{j}.png", "JPEG")

convert_pdf_to_images("./ocr.pdf")
```

```python
import pytesseract
from PIL import Image
pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'
global pages
print(pages)
def perform_ocr_on_images():
    for i in range(pages):
        print(f"Page {i + 1}:")
        for j in range(boxes):
            image_path = f"./pages/page_{i}_box_{j}.png"
            image = Image.open(image_path)
            text = pytesseract.image_to_string(image, lang="jpn")
            print(f"Box {j + 1}:\n{text}\n")

perform_ocr_on_images()
```
