```python
from pdf2image import convert_from_path
from PIL import Image
import cv2
import numpy as np
pages = 0

def cropContour(path):
    src = cv2.imread(path)
    thr = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
    _, thr = cv2.threshold(thr, 125, 255, cv2.THRESH_BINARY)
    contours, _ = cv2.findContours(thr, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)

    contours = list(contours)
    contours.sort(key=lambda contour: cv2.contourArea(contour), reverse=True)
    x, y, w, h = cv2.boundingRect(contours[1])
    a = Image.fromarray(src[y:y+h, x:x+w])
    a.save(path)

def convert_pdf_to_images(pdf_path):
    pdf = convert_from_path(pdf_path, fmt='png')
    for i in range(len(pdf)):
        global pages
        pages += 1
        page = pdf[i]
        page.save(f"/content/pages/page_{pages}.png")
        cropContour(f"/content/pages/page_{pages}.png")

convert_pdf_to_images("./ocr.pdf")
```

```python
import pytesseract
from PIL import Image
pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'

global pages
for i in range(pages):
    print(f"Page {i + 1}:")
    image = Image.open(f"/content/pages/page_{i + 1}.png")
    print(pytesseract.image_to_string(image, lang="jpn"))
```
